      Between-group variance ÷ Within-group variance
  
  2. p-value
     
     Statistical significance understand .
     p < 0.05 -> Significant difference
     p > 0.05 -> No Significant difference

  3. Degress of Freedom (df)

       Degress of divide by 2 types:-

      1. df_between = k -1

        (k = number of groups)

      2. df_within = N - k
         
        (N = total observations)
  
| Parameter          | Work                  |
| ------------------ | --------------------- |
| F-value            | Variance ratio        |
| p-value            | Significance check    |
| Degrees of Freedom | Calculation structure |

```python
# ONE WAY ANNOVA (BASIC-EXAMPLE)
# RMEMBER:- ANNOVA ONLY TELL US DIFFERENCE HAVE BUT NEVER TELL US WHICH GROUP DIFFERENT FROM ANOTHER GROUP .
from scipy.stats import f_oneway

# 3 group

Group1 = [50 ,55, 60]
Group2 = [65, 70, 75]
Group3 = [80, 85, 90]

#ANNOVA test

f_stat, p_value = f_oneway(Group1, Group2, Group3)

print("F-Statistic:", f_stat)
print("P-value:", p_value)
```

#**What is In-Sample Evaluation ?***

     The data we use and train the model , these same data use and evaluate the model .

In Simple language:-  
     Training data is a learning data, Evaluate data is a testing data.

# **`Visual Flow`**
      
      ``Raw Data``
            |
      ``X(Input), y(Output)``
            |
      ``Train/Test Split``
            |
      ``Model Train``
            |
      ``Model Test``
   

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import numpy as np

# Data
X = np.array([2,4,6,8]).reshape(-1,1)
y = np.array([40,50,70,85])

# Model
model = LinearRegression()
model.fit(X, y)

# Prediction on SAME data
y_pred = model.predict(X) #Here we using X for prediction .This is in-Sample Evaluation .

print("R2 Score:", r2_score(y, y_pred))

# What kind of miss understanding are happen ?

    # Because Model already see this data . So:- Probability very high in r2 score result .
    #Very minimum amount of amount can come . You can't detect Overfitting .
```

#``What is Out-of-Sample Evaluation ?``

```python
# Train data ≠ Test data

from sklearn.model_selection import train_test_split # This is a function that help to divide into 2 parts .Train data, Test data .

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # test_size=0.3 . Means 30% data will be used in testset and 70% data will be use train .
                                                                         #If total 10 data points have . then 3 data use in test set and 7 data will be use in training .



#       |
#      \|/
#Data split here .
#Hwat happen here ?

#X = input
#Y = Output / target

model.fit(X_train, y_train) #Model train here , Model learn only with training data . Train data has not been used yet.

y_pred = model.predict(X_test) # Now model prediction on new data in (X_test) .
y_pred

#Here model learn from new data . this is very reliable .
```
